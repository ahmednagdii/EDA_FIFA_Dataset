# Features Describtion 
* **ID:** Unique identifier for each player in the dataset.
* **Name:** Name of the player.
* **Age:** Age of the player.
* **Nationality:** Nationality of the player.
* **Overall:** Overall rating of the player's skills and abilities.
* **Potential:** Potential rating indicating how much a player's skills can improve.
* **Club:** Current club of the player.
* **Value:** Market value of the player.
* **Wage:** Weekly wage of the player.
* **Preferred Foot:** Dominant foot of the player (left or right).
* **International Reputation:** Reputation of the player on an international level.
* **Skill Moves:** Rating indicating the skill moves of the player.
* **Position:** Playing position of the player.
* **Joined:** Date when the player joined the current club.
* **Contract Valid Until:** Date until which the player's contract with the club is valid.
* **Height:** Height of the player.
* **Weight:** Weight of the player.
* **Release Clause:** Release clause value indicating the amount required to buy out the player's contract.

### Most Total value and number of players in Nationality

```python
df.groupby('Nationality').agg(
    Total_value=('Value','sum'),
    Player_count=('Name','count')).sort_values(by=['Total_value'],ascending=False)
```
